```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

dados = pd.read_csv('Aeroporto Barajas 1997-2015.csv')
dados.shape
```

```python
dados.describe()
```

```python
dados.head(10)
```

```python
print(dados.columns.values)
```

```python
dados.plot(x='Min TemperatureC', y='Max TemperatureC', style='.', color='blue')
plt.title('MÃ­nima vs MÃ¡xima Temperatura')
plt.xlabel('Temperatura MÃ­nima')
plt.ylabel('Temperatura MÃ¡xima')
plt.show()
```

```python
plt.figure()
plt.tight_layout()
plt.title('SomatÃ³rio Temperatura MÃ¡xima')
sns.distplot(dados['Max TemperatureC'], color='0')
plt.show()
```

```python
dados.isnull().sum()
```

```python
dados.dropna(axis=0, inplace=True)
```

```python
dados.isnull().sum()
```

```python
dados.shape
```

```python
X = dados['Min TemperatureC'].values.reshape(-1,1)
y = dados['Max TemperatureC'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_test.shape
```

```python
model = LinearRegression() 
model.fit(X_train, y_train)
```

```python
float(model.coef_)
```

```python
float(model.intercept_)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
dados = pd.DataFrame({'Original':y_test.flatten(), 'Prevista':y_pred.flatten()})
pd.set_option('display.max_rows', 301)
dados
```

```python
plot = dados.head(20)
plot.plot(kind='bar')
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='black')
plt.show()
```

```python
print(metrics.mean_squared_error(y_test, y_pred))
```

```python
print(metrics.mean_absolute_error(y_test, y_pred))
```
